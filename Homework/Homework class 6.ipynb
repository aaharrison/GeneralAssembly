#### Clean date data function

```python
import datetime
leapYear = np.arange(1904, 2020, 4).tolist()

def dateFixer(df):
    dateList = []
    for x in df["Date"]:
        year = int(str(x).split(".")[0])
        if year in leapYear:
            month = round((int(str(x).split(".")[1])/1000) * 12 + .5)
        else:
            month = round((int(str(x).split(".")[1])/1000) * 12 + .5)

        date = datetime.date(year,month,1)

        dateList.append(datetime.datetime.strftime(date, "%m/%Y"))
    
    df["Clean Date"] = dateList
    
    cleanDate = []
    for x in df["Clean Date"]:
        cleanDate.append(datetime.datetime.strptime(x, "%m/%Y"))
    
    df["Clean Date"] = cleanDate
```
